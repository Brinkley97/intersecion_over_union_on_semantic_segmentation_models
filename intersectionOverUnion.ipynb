# Intersection Over Union (IoU)

- Article: [IOU](https://medium.com/analytics-vidhya/iou-intersection-over-union-705a39e7acef#:~:text=IOU(Intersection%20over%20Union)%20is,fits%20perfectly%20around%20an%20object) by Vineeth S Subramanyam

```python
def iou(box1, box2):
    
    x1, y1, x2, y2 = box1
    x3, y3, x4, y4 = box2
    
    x_inter1 = max(x1, x3)
    print("x_inter1 : ", x_inter1)
    y_inter1 = max(y1, y3)
    print("y_inter1 : ", y_inter1)
    x_inter2 = max(x2, x4)
    print("x_inter2 : ", x_inter2)
    y_inter2 = max(y2, y4)
    print("y_inter2 : ", y_inter2)
    
    width_inter = abs(x_inter2 - x_inter1)
    print("\nwidth_inter : ", width_inter)
    height_inter = abs(y_inter2 - y_inter1)
    print("height_inter : ", height_inter)
    a_o_i = width_inter * height_inter
    print("a_o_i : ", a_o_i)
    
    width_box1 = abs(x2 - x1)
    print("\nwidth_box1 : ", width_box1)
    height_box1 = abs(y2 - y1)
    print("height_box1 : ", height_box1)
    width_box2 = abs(x4 - x3)
    print("width_box2 : ", width_box2)
    height_box2 = abs(y4 - y3)
    print("height_box2 : ", height_box2)
    
    area_box1 = width_box1 * height_box1
    print("\narea_box1 : ", area_box1)
    area_box2 = width_box2 * height_box2
    print("area_box2 : ", area_box2)
    area = area_box1 + area_box2
    print("\narea : ", area)
    a_o_u = area - a_o_i
    print("a_o_u : ", a_o_u)
    
    i_o_u = a_o_i / a_o_u
    return i_o_u
```

```python
b1 = [1, 2, 3, 4]
b2 = [5, 6, 7, 8]
i_o_u = iou(b1, b2)
```

# Ex with Image
- Article: [IoU calculation for evaluating an image segmentation model](https://towardsdatascience.com/intersection-over-union-iou-calculation-for-evaluating-an-image-segmentation-model-8b22e2e84686) by Oleksii Sheremet

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def iou_with_image(image):
    first_bb_points = [
                  [400, 300], 
                  [600, 0],
                  [700, 320], 
                  [600, 320]
                  ]
    stencil = np.zeros(img.shape).astype(img.dtype)
    contours = [np.array(first_bb_points)]
    color = [255, 255, 255]
    cv2.fillPoly(stencil, contours, color)
    result1 = cv2.bitwise_and(img, stencil)
    result1 = cv2.cvtColor(result1, cv2.COLOR_BGR2RGB)
    # plt.imshow(result1)
    
    
    second_bb_points = [
                  [400, 300], 
                  [600, 0],
                  [700, 320], 
                  [600, 320]
                  ]
    stencil2 = np.zeros(img.shape).astype(img.dtype)
    contours2 = [np.array(second_bb_points)]
    color = [255, 255, 255]
    cv2.fillPoly(stencil2, contours2, color)
    result2 = cv2.bitwise_and(img, stencil2)
    result2 = cv2.cvtColor(result2, cv2.COLOR_BGR2RGB)
    # plt.imshow(result1)


    intersection = np.logical_and(result1, result2)
    union = np.logical_or(result1, result2)
    iou_score = np.sum(intersection) / np.sum(union)
    return "IoU is %s" % iou_score
```
